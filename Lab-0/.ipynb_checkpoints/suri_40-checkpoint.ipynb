```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
# A simple dataframe with the courses I am taking and their corresponding credit hours / week
courses = pd.DataFrame({'credit_hr_week': [3, 3, 3, 2]},
                  index=['Data_Mining', 'Neural_Networks', 'Algorithms', 'Artificial_Intelligence'])
```

```python
courses.head()
```

```python
# Pie chart showing the courses I am taking with their corresponding credit hours per week
# to give an estimate of workload in this semester.
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot()
explode = (0.1, 0, 0, 0)
plot = ax.pie(courses['credit_hr_week'],
              explode=explode,
              labels=courses.index,
              autopct='%1.1f%%',
              shadow=True)
legend = ax.legend(loc='best')
plt.show()
```
