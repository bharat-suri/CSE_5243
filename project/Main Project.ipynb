```python
import pandas as pd
```

```python
df=pd.read_pickle('Restaurants_clean.pkl')
```

```python
for i in df.columns:
    print(i + ": " + str(type(df[i].iloc[0])))
```

```python
#Count of sparseness of each column
attr_count={}
for i in df.columns:
    attr_count[i]=df[i].astype(bool).sum(axis=0)
    #print(i + ": " + str(df[i].astype(bool).sum(axis=0)))
sr = pd.Series(attr_count)
sr.sort_values(inplace=True,ascending=False)
for ind, val in sr.iteritems():
    print(ind, ": ", val)
```

```python
#Count of null values in each column - Nan
for i in df.columns:
    print(i + ": " + str(df[i].isna().sum()))
```
