```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

%matplotlib inline

sns.set(style="whitegrid")
```

## Loading and Visualizing the data

```python
df = pd.read_csv('data/lab2.csv')
df.head()
```

```python
df.describe()
```

```python
sns.distplot(df['a1'], color='maroon');
```

```python
sns.distplot(df['a2'], color='green');
```

Basic statistics summary and the plots show that the data is uniform.

```python
fig = plt.figure(figsize=[12, 8])
t = plt.title('Original dataset (True Classes)', fontsize=20)
x = plt.xlabel('a1', fontsize=15)
y = plt.ylabel('a2', fontsize=15)
s = sns.scatterplot(x='a1',
                y='a2',
                data=df,
                hue='class',
                s=50,
                palette=['maroon', 'green'])
l = plt.legend(fontsize=16)

plt.savefig('dataset.png')
plt.show()
```

The scatter plot revealed that the two classes had a soft intersection. This also means that the two classes do not have a clear linear boundary that separates them perfectly. There is, however, somewhat of a separation near the a2=-2 line that separates the two classes.

```python
data = df['class'].value_counts()
t = plt.title('Number of instances for each class', fontsize=20)
x = plt.xlabel('Class', fontsize=15)
y = plt.ylabel('Value counts', fontsize=15)
b = sns.barplot(x=data.index, y=data.values,
            palette=['maroon', 'green'])

plt.savefig('value_counts.png')
plt.show()
```

In this synthetic dataset, the two features are normally distributed and the number of instances of each class is equal.

## Preparing the data

```python
X = df.loc[:, 'a1':'a2'].values
y = df.loc[:, 'class'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

The training set now contains 350 out of the 500 original data points. The remaning 150 data points are in the testing set.

## Initializing and training the classifier

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
train = pd.Series(y_train, dtype=np.int64)
test = pd.Series(y_test, dtype=np.int64)
train_pred = pd.Series(knn.predict(X_train), dtype=np.int64)
test_pred = pd.Series(knn.predict(X_test), dtype=np.int64)
data = pd.DataFrame(X_test, columns=['a1', 'a2'])
```

```python
print('Score on training data: {:.2f}'.format(knn.score(X_train, y_train)))
```

## Testing the classifier

```python
data['test'] = test
data['test_pred'] = test_pred
```

```python
fig = plt.figure(figsize=[17, 7])

p = plt.subplot(1, 2, 1)
t = plt.title('True classes', fontsize=20)
s = sns.scatterplot(x='a1', y='a2', hue='test', data=data, s=65,
                palette=['maroon', 'green'])
l = plt.legend(loc=1, fontsize=12)
x = plt.xlabel('a1', fontsize=15)
y = plt.ylabel('a2', fontsize=15)

p = plt.subplot(1, 2, 2)
t = plt.title('Predicted classes', fontsize=20)
s = sns.scatterplot(x='a1', y='a2', hue='test_pred', data=data, s=65,
                palette=['maroon', 'green'])
l = plt.legend(loc=1, fontsize=12)
x = plt.xlabel('a1', fontsize=15)
y = plt.ylabel('a2', fontsize=15)

plt.savefig('final.png')
plt.show()
```

The miscalssified points are the ones that have more points of the opposite class nearer to them.

```python
print('Accuracy: {:.2f}'.format(knn.score(X_test, y_test)))
```

```python
# Varying the value of k to see the effect on train and test set

train = []
test = []
n = np.arange(3, 40, 2)
for k in n:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train.append(knn.score(X_train, y_train))
    test.append(knn.score(X_test, y_test))
```

```python
fig = plt.figure(figsize=(12, 8))
l1 = sns.lineplot(n, train, marker='o')
l2 = sns.lineplot(n, test, marker='o')
l = plt.legend(['train', 'test'], loc=1, fontsize=15)
x = plt.xlabel('Value of k', fontsize=16)
y = plt.ylabel('Score', fontsize=16)
t = plt.title('Score for train and test sets with various k values', fontsize=20)
```

The above plot shows that for the given test/train split, score on test set is always lower than the train set for different values of k.
